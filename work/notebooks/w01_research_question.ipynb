## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision: Which pages should the content team refresh this week?

Action: A content writer picks the top-ranked pages from
the model's output and rewrites/updates them first.

Cost of a wrong recommendation:
- False positive: Writer wastes time refreshing a healthy
  page that didn't need work.
- False negative: A truly declining page gets ignored,
  keeps losing traffic, and the client loses revenue.

The cost of missing a declining page is higher than
refreshing a healthy one — so recall matters more than
precision here.
